```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
prof_2 = pd.read_csv('profiling_2.csv')
prof_2.head()
```

```python
prof_2.copy()
prof_2 = prof_2.drop(columns=['Cluster'], axis = 1)
prof_2.index = ['Cluster 1','Cluster 2']
prof_2.head()
```

```python
plt.figure(figsize=(20, 6))
sns.heatmap(prof_2, annot=True, cmap='crest', fmt='.2f', cbar = True, xticklabels=True, yticklabels=True)

plt.show()
```
